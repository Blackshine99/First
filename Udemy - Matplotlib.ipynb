```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.plot(x,y)
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
ax1.plot(x,y)
ax2 = fig1.add_axes([0.2,0.5,.2,.2])
ax2.plot(x,y)
```

```python
fig2 = plt.figure()
ax1 = fig2.add_axes([0,0,1,1])
ax2 = fig2.add_axes([0.2,0.5,0.4,0.4])
ax1.plot(x,z)
ax2.plot(x,y)
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)


```

```python
fig3,axes = plt.subplots(nrows=1, ncols=2,figsize = (12,2))
axes[0].plot(x,y,lw = 3,color = 'b',ls = '--')
axes[1].plot(x,z,lw = 3, color = 'r')
```
